```python
import numpy as np #for numerical operations
import matplotlib.pyplot as plt # to generate plots
```

```python
#linearly spaced array
x = np.linspace(-10,10,20) #20 points between -10 and 10
```

```python
x
```

```python
m = 50 #slope
b = 17 #the b interecpet in: y=mx+b
```

```python
y = m * x + b #the equation for the line
```

```python
plt.plot(x,y,'.') #plotting the line
```

```python
r = 100*np.random.randn(20,) #to genrate random noise for the y values
```

```python
r
```

```python
plt.plot(r,'.') #to plot the noisy data
```

```python
y2 = m * x + b + r #adding the noise to the og line equation to demonstrate real data
```

```python
plt.plot(x,y2,'.') # plot of noisy data with respect to the og line
```

```python
m_guess = 1 #initial guess for slope
b_guess = 1 #initial guess for b interecpt

y_guess = m_guess * x + b_guess #to generate the line using m and b current guesses

plt.plot(x,y_guess,'r.'); #plotting guess line
plt.plot(x,y2,'g.'); #plotting noisy data (again)
```

```python
#loop thru different slope values to find best fit line
for m_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x #to generate the line using current m guess

        plt.plot(x,y_guess,'r.'); #plotting guess line
        plt.plot(x,y2,'g.'); #plotting noisy data
        plt.show()

        #calculating the loss between guess line and noisy data
        loss = np.sum((y_guess - y)**2) #sum of squared differences

        print(loss,'M = ',m_guess) #print the loss and current m guess
```

```python
#initializing variables to store; best loss, corresponding m (i.e. slope), corresponding intercept (i.e. b)
best_loss = 100000000000
best_m = 0
best_b = 0

#looping thru different combinations of slope and intercept to find best fit line
for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess #generating the line using m guess and b guess

        loss = np.sum((y_guess - y)**2)

       #updating best loss and corresponding slope and interecpt if current loss is smaller
        if loss < best_loss:

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
```

```python
#best fit line
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.'); #best fit line
plt.plot(x,y2,'g.'); #noisy data
plt.show()

print(best_loss,best_m,best_b) #printing best loss and coressponding slope and intercept
```

```python
#Set up varaibles to track the Best Fit line
best_loss = 1000000000
best_m = 0
best_b = 0

#loop through possible m and b values to find best fit line
for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        print("Now Testing: ", m_guess, b_guess) #current test values

        y_guess = m_guess * x + b_guess #y value of current guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_loss = loss #update best loss
            best_m = m_guess #update best slope
            best_b = b_guess #update best intercept
```

```python
#data points
X = np.array([19,65,66,80,97,113,175,270,310,350,350,650,680])
Y = np.array([1150,3800,5000,4800,5400,6700,10300,15400,19500,23200,21600,39400,41000])
```

```python
plt.plot(X,Y,'g.') #plot data points in green dots
```

```python
#initial guesses for slope and intercept
m_guess = 0
b_guess = 0
```

$y=mx+b$

```python
m_guess * X + b_guess
```

```python
y_guess - Y #difference between guessed value and actual value
```

```python
 #calculating 'loss' between the guessed and actual y values
np.sum((y_guess - Y)**2)
loss = np.sum((y_guess - Y)**2)
```

```python
#set up values to track best fit line
best_loss_so_far = 1000000000000
m_best = 0
b_best = 0
#loop through possible m and b values to find the best fit line
for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * X + b_guess #y value of current m and b guess

        loss = np.sum((y_guess - Y)**2) #loss = current guess and actual y values

       #update best fit line if the current loss is smaller than the previous best loss
        if loss < best_loss_so_far:
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss

print("Found best fit:")
print(m_best, b_best, loss)
```

```python
H = m_best #assigning best slope to varaible H (Ho is 'Hubble's constant')
```

If the universe has been expanding since its beginning at a constant speed, the universe's age would simply be 1/Ho.
Find the inverse of your value of Ho.
Multiply the inverse by $3.09 \times 10^{19}$ km/Mpc to cancel the distance units.
Since you now have the age of the Universe in seconds, divide this number by the number of seconds in a year: $3.16 \times 10^{7}$ sec/yr.

$3.09 \times 10^{19}$ km/Mpc

$3.16 \times 10^{7}$ sec/year

```python
(1/H) * (3.09*10**19) #estimated age of the universe in seconds
```

```python
(1/H) * (3.09*10**19)  / (3.16*10**7) #convert estimated age of universe to years
```

```python
(1/H) * (3.09*10**19)  / (3.16*10**7) / (1000000000) #convert estimated age of universe to billions of years
```

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#defining a function to fetch data from Google Sheets using URL and Sheet name
def get_database(url,sheet_name):
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```

```python
url = 'https://docs.google.com/spreadsheets/d/1B_HW2uVGQv9W7b6vVZDSkOPLtmfodHrX4G0qtQbIenk/edit?usp=sharing'
```
